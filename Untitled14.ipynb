```python
import yfinance as yf
from darts import TimeSeries
from darts.models import RNNModel
from darts.dataprocessing.transformers import Scaler
from darts.utils.missing_values import fill_missing_values
import pandas as pd

# Download historical data using yfinance
goog = yf.Ticker('GOOG')
data = goog.history(start='2023-01-01', end='2024-12-31')

# data = fill_missing_values(TimeSeries.from_dataframe(data,value_cols=["Close"]),"auto",fill_missing_dates=True)

# Create a TimeSeries from the downloaded data
series = TimeSeries.from_dataframe(data,value_cols=['Close'],fill_missing_dates=True, freq='D')

# Initialize the Darts scaler
scaler = Scaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(series)

# Split the dataset into training and test sets
train, test = series.split_after(pd.Timestamp('2024-01-01'))

# Initialize the RNN model
rnn_model = RNNModel(input_chunk_length=10, output_chunk_length=1, n_epochs=100)

# Fit the model on the training data
rnn_model.fit(train)

# Make predictions on the test dataset
prediction = rnn_model.predict(len(test))

# Inverse transform the predictions (if needed)
inverted_prediction = scaler.inverse_transform(prediction)

# Plot the results or evaluate the model
# (You can use matplotlib for visualization)
```
